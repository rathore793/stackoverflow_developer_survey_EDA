`employment(string)` function take the string as input and check for keywords and return the employment type based on the keyword

```python
#Function to get the employment type
def employment(string):
    if 'Employed, full-time' in string:
        return 'Employed, full-time'
    elif 'Employed, part-time' in string:
        return 'Employed, part-time'
    elif 'Independent contractor, freelancer, or self-employed' in string:
        return 'Independent contractor, freelancer, or self-employed'
    elif 'Not employed, but looking for work' in string:
        return 'Not employed, but looking for work'
    elif 'Not employed, and not looking for work' in string:
        return 'Not employed, and not looking for work'
    elif 'Student, full-time' in string:
        return 'Student, full-time'
    elif 'Student, part-time' in string:
        return 'Student, part-time'
    elif 'Retired' in string:
        return 'Retired'
    else:
        return 'Not Disclosed'
    
```

```python
new_df['Employment'] = new_df['Employment'].apply(employment)
```

```python
new_df.head(3)
```

```python
#Checking all eduction level values
new_df['EdLevel'].value_counts()
```

```python
#Replacing the old values with the simple ones
new_df['EdLevel'].replace(to_replace=['Bachelor’s degree (B.A., B.S., B.Eng., etc.)', 'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)', 'Some college/university study without earning a degree', 'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)', 'Professional degree (JD, MD, Ph.D, Ed.D, etc.)', 'Associate degree (A.A., A.S., etc.)', 'Primary/elementary school'], 
                        value=["Bachelor's Degree", "Master's Degree", "College but no degree", "Secondary School", "Professional Degree", "Associate Degree", "Primary School"], inplace=True)
```

`learning(string)` function to get the learning source of coding. Function take string as input which is number of diffrent options based on that we can get the most useful or common source of learning

```python
#Function to get leanring source of code
def learning(string):
    if 'Books / Physical media' in string:
        return 'Books / Physical media'
    elif 'Coding Bootcamp' in string:
        return 'Coding Bootcamp'
    elif 'Colleague' in string:
        return 'Colleague'
    elif 'Friend or family member' in string:
        return 'Friend or family member'
    elif 'Hackathons' in string:
        return 'Hackathons'
    elif 'Online Courses or Certification' in string:
        return 'Online Courses or Certification'
    elif 'On the job training' in string:
        return 'Job training'
    elif 'Other online resources (e.g., videos, blogs, forum)' in string:
        return 'Other online resources (e.g., videos, blogs, forum)'
    elif 'School (i.e., University, College, etc)' in string:
        return 'School or College'
    else:
        return 'Other'
```

```python
#Saving data into the same column
new_df['Learning Source'] = new_df['LearnCode'].apply(learning)
```

### Q1 - What Companies do the Developer get paid most?

*Basis below we can clearly see the in companies like Advertising Services, Insurance, Healthcare and Financial Services the developers are paid most.*

### Q2 - How much remote working matters to employee?

```python
final_df['RemoteWork'].value_counts()
```

*Majority of the People have been doing Remote work or Hybrid, also the average pay scale of these two is higher than In-person, hence people prefer Remote working more than In-person*

```python
final_df[final_df['RemoteWork'] != 'N/A'].groupby('RemoteWork').agg(np.mean).sort_values(by='Income_in_$')
```

### Q3 - How does coding experience affects level of pay?

*People with more coding experience get paid more, as per below 36-50 group get the highest salary among all group with average person have coding experience of 40.8 years*
